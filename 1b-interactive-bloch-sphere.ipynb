```python
import numpy as np
from qiskit.visualization.bloch import Bloch
import ipywidgets as widgets
from IPython.display import display, clear_output

def statevector_to_bloch(statevec):
    alpha = statevec[0]
    beta = statevec[1]
    x = 2 * (alpha.conjugate() * beta).real
    y = 2 * (alpha.conjugate() * beta).imag
    z = abs(alpha)**2 - abs(beta)**2
    return [x, y, z]
```

# Rotations with $\theta$ and $\phi$

## $$\ket{\psi} = \cos\frac{\theta}{2} \, \ket{0} + e^{i \phi} \, \sin\frac{\theta}{2} \, \ket{1}$$

Two sliders representing the polar angle (θ) and azimuthal angle (φ) of a single qubit.  
- **θ slider:** rotates the qubit state from the north pole (|0⟩) toward the south pole (|1⟩).  
- **φ slider:** rotates the state around the z-axis, controlling the relative phase.

---

```python
def bloch_from_angles(theta, phi):
    # General qubit state:
    # |ψ⟩ = cos(θ/2)|0⟩ + e^{iφ} sin(θ/2)|1⟩
    alpha = np.cos(theta / 2)
    beta = np.exp(1j * phi) * np.sin(theta / 2)
    return np.array([alpha, beta])

def update_1(theta, phi):
    clear_output(wait=True)
    
    state = bloch_from_angles(theta, phi)
    bloch_vec = statevector_to_bloch(state)

    b = Bloch()
    b.add_vectors(bloch_vec)
    b.vector_color = ["r"]
    
    b.show()
```

```python
theta_slider = widgets.FloatSlider(
    min=0, max=np.pi, step=0.01, value=0, description='θ'
)
phi_slider = widgets.FloatSlider(
    min=0, max=2*np.pi, step=0.01, value=0, description='φ'
)
```

```python
ui = widgets.VBox([theta_slider, phi_slider])
out = widgets.interactive_output(update_1, {'theta': theta_slider, 'phi': phi_slider})

display(ui, out)
```

# Rz, Rx, Ry

Three sliders controlling rotations around the X, Y, and Z axes.  
- **Rx slider:** rotates the qubit about the x-axis.  
- **Ry slider:** rotates the qubit about the y-axis.  
- **Rz slider:** rotates the qubit about the z-axis.  
These rotations are applied in a fixed order to produce the resulting Bloch vector.

```python
# Rotation matrices
def Rx(theta):
    return np.array([[np.cos(theta/2), -1j*np.sin(theta/2)],
                     [-1j*np.sin(theta/2), np.cos(theta/2)]])

def Ry(theta):
    return np.array([[np.cos(theta/2), -np.sin(theta/2)],
                     [np.sin(theta/2), np.cos(theta/2)]])

def Rz(theta):
    return np.array([[np.exp(-1j*theta/2), 0],
                     [0, np.exp(1j*theta/2)]])
```

```python
def update_2(rx, ry, rz):
    clear_output(wait=True)
    
    # Start in |0⟩
    state = np.array([1, 0], dtype=complex)
    
    # Apply rotations Rx, Ry, Rz in order
    state = Rz(rz) @ Ry(ry) @ Rx(rx) @ state
    
    bloch_vec = statevector_to_bloch(state)
    
    b = Bloch()
    b.add_vectors(bloch_vec)
    b.vector_color = ["r"]
    
    b.show()
```

```python
rx_slider = widgets.FloatSlider(min=0, max=2*np.pi, step=0.01, value=0, description='Rx')
ry_slider = widgets.FloatSlider(min=0, max=2*np.pi, step=0.01, value=0, description='Ry')
rz_slider = widgets.FloatSlider(min=0, max=2*np.pi, step=0.01, value=0, description='Rz')
```

```python
ui = widgets.VBox([rx_slider, ry_slider, rz_slider])
out = widgets.interactive_output(update_2, {'rx': rx_slider, 'ry': ry_slider, 'rz': rz_slider})

display(ui, out)
```

# Swapping order of operations

Two sliders for rotations (Rx and Rz) and a checkbox to swap their order.  
- **Rx slider:** controls rotation around the x-axis.  
- **Rz slider:** controls rotation around the z-axis.  
- **Swap Order checkbox:** toggles the order of applying Rx and Rz to demonstrate that rotation order affects the final state.

```python
def update_3(rx_angle, rz_angle, swap_order):
    clear_output(wait=True)
    state = np.array([1,0], dtype=complex)
    
    # Apply rotations in chosen order
    if swap_order:
        state = Rx(rx_angle) @ Rz(rz_angle) @ state
    else:
        state = Rz(rz_angle) @ Rx(rx_angle) @ state
    
    bloch_vec = statevector_to_bloch(state)
    
    b = Bloch()
    b.add_vectors(bloch_vec)
    b.vector_color = ["b"]
    b.show()
```

```python
rx_slider = widgets.FloatSlider(min=0, max=2*np.pi, step=0.01, value=np.pi/2, description='Rx')
rz_slider = widgets.FloatSlider(min=0, max=2*np.pi, step=0.01, value=np.pi/2, description='Rz')
swap_checkbox = widgets.Checkbox(value=False, description='Swap Order (Rx↔Rz)')
```

```python
ui = widgets.VBox([rx_slider, rz_slider, swap_checkbox])
out = widgets.interactive_output(update_3, {'rx_angle': rx_slider, 'rz_angle': rz_slider, 'swap_order': swap_checkbox})

display(ui, out)
```

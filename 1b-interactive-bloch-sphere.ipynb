# Rotations with $\theta$ and $\phi$

## $$\ket{\psi} = \cos\frac{\theta}{2} \, \ket{0} + e^{i \phi} \, \sin\frac{\theta}{2} \, \ket{1}$$

Two sliders representing the polar angle (θ) and azimuthal angle (φ) of a single qubit.  
- **θ slider:** rotates the qubit state from the north pole (|0⟩) toward the south pole (|1⟩).  
- **φ slider:** rotates the state around the z-axis, controlling the relative phase.

---

# Rz, Rx, Ry

Three sliders controlling rotations around the X, Y, and Z axes.  
- **Rx slider:** rotates the qubit about the x-axis.  
- **Ry slider:** rotates the qubit about the y-axis.  
- **Rz slider:** rotates the qubit about the z-axis.  
These rotations are applied in a fixed order to produce the resulting Bloch vector.

# Swapping order of operations

Two sliders for rotations (Rx and Rz) and a checkbox to swap their order.  
- **Rx slider:** controls rotation around the x-axis.  
- **Rz slider:** controls rotation around the z-axis.  
- **Swap Order checkbox:** toggles the order of applying Rx and Rz to demonstrate that rotation order affects the final state.
